```python
import pandas as pd
df = pd.read_csv("Resources/Week3-Day2-Activity8-annual_gold.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
df.dtypes
```

```python
df.plot()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
plt.figure(figsize=(20,10))
x = df['Date']
y = df['Price']
plt.plot(x,y)
plt.title('Price of Golds',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(20,10))
x = df['Date']
y = df['Price']
plt.bar(x,y)
plt.title('Price of Golds',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(20,10))
x = df['Date']
y = df['Price']
plt.scatter(x,y)
plt.title('Price of Golds',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Price')
```
